```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
read_data = pd.read_csv('ds_salaries.csv')
```

```python
# Print the first five rows of our Dataset
```

```python
# What is the highest salary job title in the field of Data science and what is the job title?
```

```python
max_salary = read_data['salary_in_usd'].max()
print("The highest salary for a job in the field of data science is: {}".format(max_salary))
```

```python
# What is the minimum salary in USD?
```

```python
min_salary = read_data['salary_in_usd'].min()
print("The lowest salary for a job in the field of data science is: {}".format(min_salary))
```

```python
mean_salary = read_data['salary_in_usd'].mean()
print("The average salary for data science job is: {}".format(mean_salary))
```

```python
read_data.sort_values('salary_in_usd', ascending = False)
```

```python
# make a graph of the salary in usa dollar according to the job title
plt.rcParams['figure.figsize'] = (25,15)
```

```python
plt.plot(read_data.salary_in_usd, read_data.job_title, 'o')
```

```python
# plot a bar chart for the same data; salary as y axis and job as x axis
```

```python
read_data.plot.bar(x = 'job_title', y = 'salary_in_usd', rot=0, color = "blue", width = 50)
```
